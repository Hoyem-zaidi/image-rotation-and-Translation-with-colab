```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
#rotate image 90°
img = cv2.imread("/content/sample.jpg")
img = cv2.resize(img, (img.shape[1]//2, img.shape[0]//2))
plt.subplot(2,1,1),plt.imshow(img)
plt.subplot(2,1,2),plt.imshow(np.rot90(img))
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('/content/sample.jpg')

# Store height and width of the image
height, width = image.shape[:2]

quarter_height, quarter_width = height / 3, width / 3

T = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]])
# image Translation 
# We use warpAffine to transform
# the image using the matrix, T
img_translation = cv2.warpAffine(image, T, (width, height))
# display image
plt.figure()
plt.subplot(2,1,1),plt.imshow(image)
plt.subplot(2,1,2),plt.imshow(img_translation)
plt.show()
```
